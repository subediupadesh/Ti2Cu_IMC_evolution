```python
import netCDF4
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from natsort import natsorted

import subprocess
from matplotlib.animation import FuncAnimation

from matplotlib.colors import ListedColormap
from scipy.io import loadmat
import natsort, pathlib
import os
import glob

```

```python
time = pd.read_csv('temp_csv/time_calculator_r2.csv')['t(s)']
time.tail()
```

## Interface Start

```python
path_1 = "temp_csv/interface_start/"
csv_files_1 = natsorted(glob.glob(path_1+'/*.csv'))  # Importing csv files of simulation naturally sorted
```

```python
start_int = np.zeros((len(csv_files_1), 2))
for i, file in enumerate(csv_files_1):
    df1 = pd.read_csv(file, index_col=None)
    start_int[i][0] = df1['temperature']
    start_int[i][1] = time[i]
```
